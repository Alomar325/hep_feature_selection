## RevisiÃ³n de atributos faltantes (missing attributes)

```python
def show_null(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    display(missing_data)
show_null(df)
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, cmap='BrBG', annot = True, fmt=".2f", annot_kws={"size":7})
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':8}, pad=12)
```

## GrÃ¡fico de puntos entre pares de variables

```python
plt.figure(figsize=(7, 4))
plt.scatter(df["PRI_jet_all_pt"], df["PRI_tau_pt"],s=0.5, alpha=0.6,color="C0")
plt.grid(ls=":", color="k")
plt.xlabel("lepton_pt")
plt.ylabel("lepton_eta")
plt.show()
```
